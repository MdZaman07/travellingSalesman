### Search problem: TSP (travelling salesman problem)
Suppose a travelling agent travels in a region, say Sydney. The locations of cities (e.g. the cities Sydney) are known. Every city is connected with other cities via paths in both directions. The agent's task is to start from a city (called the start city which is given), visit every other city only once and lastly return to the start city with the mininal travel cost or the shortest path. 

Create a dictionary of 20 locations for our Travelling Salesman Problem

```python
locations = {"Bankstown":(33.9178,151.0264), "Barangaroo":(33.8638,151.2015), "Blacktown":(33.7661,150.9127),
             "Bondi":(33.8915,151.2767), "Central":(33.7812,151.9665), "Chatswood":(33.7961,151.1780), 
             "Epping":(33.7739,151.0831),"Enfield":(33.8894,151.0953), "Granville":(33.8355,151.0095), 
             "Hurstville":(33.9646,151.1010), "Kingsgrove":(33.9379,151.0995),
             "Lidcombe":(33.8617,151.0423), "Liverpool":(33.9129,260.7447),"Parramatta":(33.8148,151.0017),
             "Penrith":(33.7507,150.6877), "Punchbowl":(33.9325,151.0449), "Redfern":(33.8930,151.2054),
             "Riverwood":(33.9454,151.0534), "Ryde":(33.8130,151.1050), "Strathfield":(33.8774,151.0824)}
```

Create function define_map which takes the dictionary(locations) as parameter and returns the list of all cities and a dictionary. The disctionary is nested with cities as keys and values representing another dictionary with cities as keys and values are the distances between the cities in the outer dictionary key and cities in the inner dictionary key. 

```python
def define_map(city_location):
    global distances
    city_locations = city_location
    all_cities = []
    distances = {}

    for city in city_locations.keys():
        distances[city] = {}
        all_cities.append(city)
    all_cities.sort()

    for name_1, coordinates_1 in city_locations.items():
            for name_2, coordinates_2 in city_locations.items():
                distances[name_1][name_2] = np.linalg.norm(
                    [coordinates_1[0] - coordinates_2[0], coordinates_1[1] - coordinates_2[1]])
                distances[name_2][name_1] = np.linalg.norm(
                    [coordinates_1[0] - coordinates_2[0], coordinates_1[1] - coordinates_2[1]])

    return (all_cities, distances)
```

We use the dictionary of cities as input into the function so that each distance can be calculated using the straight line algorithm. 

```python
all_cities, distances = define_map(locations)
```

```python
def neighbour_for_TSP(state):
    neighbour_state = state[:]
    left = random.randint(1, len(neighbour_state) - 1)
    right = random.randint(1, len(neighbour_state) - 1)
    if left > right:
        left, right = right, left
    neighbour_state[left: right + 1] = reversed(neighbour_state[left: right + 1])
    return neighbour_state
```

Function above takes a state and returns the states on its left and right, which are the neighbours of it. 

```python
def cost_fun(state):
    cost = 0
    for i in range(len(state) - 1):
        cost += distances[state[i]][state[i + 1]]
    cost += distances[state[0]][state[-1]]
    return cost
```

The cost function above calculates the distance of the path from the starting city, travelling through all the cities in the state and ending with the starting city.

We take 'Strathfield' as the starting city. Any other city from the locations dictionary can be used as well

```python
start_point = 'Strathfield'
```

```python
def define_init(city_list, start_point):
    init_citylist = city_list[:]
    init_citylist.remove(start_point)
    init_citylist.insert(0, start_point)
    init_citylist.insert(len(city_list), start_point)
    return init_citylist
```

The function above takes in the city list and the starting point to create an initial solution covering all the cities, starting and ending with the start city 

We hold the define_init(all_cities,start_point) function's return value in a list called init. This is the initial path.

```python
init = define_init(all_cities, start_point)
```
